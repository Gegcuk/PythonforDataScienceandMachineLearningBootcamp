# Seaborn

```python
import seaborn as sn
%matplotlib inline
tips = sn.load_dataset('tips')
flights = sn.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tips['total_bill'].corr(tips['tip'])
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sn.heatmap(fp, cmap='coolwarm', linecolor='white', linewidths=0.01)
```

```python
sn.clustermap(fp, cmap='coolwarm', standard_scale = 1)
```

# Grids

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
iris = sn.load_dataset('iris')
```

```python
sn.pairplot(iris)
```

```python
g = sn.PairGrid(iris)
g.map_diag(sn.histplot)
g.map_upper(plt.scatter)
g.map_lower(sn.kdeplot)
```

```python
tips = sn.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sn.FacetGrid(data=tips, col='time', row='smoker')
g.map(sn.scatterplot, 'total_bill', 'tip')
```
