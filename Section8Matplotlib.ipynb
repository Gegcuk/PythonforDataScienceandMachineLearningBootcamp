```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')


plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set(title='Title')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)
plt.tight_layout()
axes[0].plot(x, y)
axes[0].set_title('First plot')

axes[1].plot(y, x)
axes[1].set_title('Second plot')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=300)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Title")
ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label='x cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()\

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='g', lw=3, ls=':', marker='8', markersize = 10, markerfacecolor='g',
        markeredgewidth=2, markeredgecolor='y')

ax.set_xlim([1, 10])
ax.set_ylim([1, 30])
```
